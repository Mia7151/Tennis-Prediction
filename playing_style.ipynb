```python
from lxml import etree
import urllib.request
import urllib.parse
import pandas as pd
import unicodedata
```

```python
def query(content):
    # 请求地址
    url = content
    # 请求头部
    headers = { 
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36' 
    }
    # 利用请求地址和请求头部构造请求对象
    req = urllib.request.Request(url = url, headers = headers, method = 'GET')
    # 发送请求，获得响应
    try:
        response = urllib.request.urlopen(req)
        # 读取响应，获得文本
        text = response.read().decode('utf-8')
        # 构造 _Element 对象
        html = etree.HTML(text)
        # 使用 xpath 匹配数据，得到 <div class="mw-parser-output"> 下所有的子节点对象
        obj_list = html.xpath('//div[@class="mw-parser-output"]/*')
        
        #获取<h2>节点对象
        find = False
        start_h2 = 0
        end_h2 = 0
        for i in range(len(obj_list)):
            if 'h2'==obj_list[i].tag:
                if obj_list[i].xpath('.//text()')[0] == 'Playing style':
                    find = True
                    start_h2 = i+1
                    continue
            if 'h2'==obj_list[i].tag and find:
                    end_h2 = i-1
            if end_h2!=0:
                break

        if not find:
            return ""
        playing_style_p = obj_list[start_h2:end_h2+1]
        playing_style_p_list = [obj.xpath('.//text()') for obj in playing_style_p]
        playing_style = [p.encode('utf-8').decode() for playing_style in playing_style_p_list for p in playing_style]
        playing_style_after_filter = [item.strip('\n') for item in playing_style]
        
        
        return ''.join(playing_style_after_filter)
    except:
        pass
        
```
