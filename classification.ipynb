```python
import xgboost as xgb
import os
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import BernoulliNB
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
from sklearn.model_selection import cross_val_score
from scipy import stats
```

```python
def train_models(X_train,y_train, model):
    if model=='XGBoost':
        clf = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
        clf.fit(X_train,y_train)
        return clf
    
    if model=='naive_bayes':
        clf = BernoulliNB()
        clf.fit(X_train, y_train)
        return clf
    
    # if model=="logistic_regression":
    #     clf = LogisticRegression(random_state=0).fit(X_train, y_train)
    #     return clf
    
    if model=='svm':
        clf = svm.SVC()
        clf.fit(X_train, y_train)
        return clf
    
    if model=='random_forest':
        clf = RandomForestClassifier(max_depth=2, random_state=0)
        clf.fit(X_train, y_train)
        return clf
```

```python
def evaluate_model(X_val, y_val, model):
    #y_val = np.array(y_val)
    y_predict = model.predict(X_val)
    #y_predict = np.argmax(y_predict, axis=0)
    ac_score=sklearn.metrics.accuracy_score(y_val, y_predict)
    #roc_auc_macro=sklearn.metrics.roc_auc_score(y_val, y_predict, multi_class="ovr")
    #roc_auc_micro=sklearn.metrics.roc_auc_score(y_val, y_predict, average='micro')
    #F1_micro=sklearn.metrics.f1_score(y_val, y_predict, average='micro')
    #F1_macro=sklearn.metrics.f1_score(y_val, y_predict, average='macro')
    
    return ac_score
```
